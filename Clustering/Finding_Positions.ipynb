```python
import pandas as pd
```

```python
names = pd.read_csv('names.csv')
pos = pd.read_csv('positions.csv')
pos.head
unique_pos = pos['positions'].unique()
list(unique_pos)
replace_dict = dict.fromkeys(list(unique_pos),None)
```

```python
replace_dict['SHOOTING GUARD'] = 'G'
replace_dict['POINT GUARD'] = 'G'
replace_dict['POWER FORWARD'] = 'C-F'
replace_dict['CENTER'] = 'C'
replace_dict['SMALL FORWARD'] = 'F'
```

```python
pos['positions'] = pos['positions'].replace(replace_dict)
```

```python
pos.head()
```

```python
names.head()
```

```python
left_join_result=pd.merge(names,pos,how='left',left_on='name_in_db',right_on='name')
left_join_result = left_join_result[['name_in_db','positions']]
left_join_result.head()
```

```python
left_join_result.to_csv('manual_work.csv',index=False)
```
