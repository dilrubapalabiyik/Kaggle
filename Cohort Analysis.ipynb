```python
def quarter_months(grouped_df):
    #dict_month={"1":"jan","2":"feb","3":"march","4":"april","5":"may","6":"june","7":"july",
                #"8":"aug","9":"sep","10":"oct","11":"nov","12":"dec"}
    first_qt=[1,2,3]
    second_qt=[4,5,6]
    third_qt=[7,8,9]
    forth_qt=[10,11,12]
    list_d=[]
    for i in range(0,len(grouped_df)):
        if grouped_df.iloc[i]['Month'] in first_qt:
            x ="Q1"
            list_d.append(x)
        elif grouped_df.iloc[i]['Month'] in second_qt:
            y ="Q2"
            list_d.append(y)
        elif grouped_df.iloc[i]['Month'] in third_qt:
            z ="Q3"
            list_d.append(z)
        else:
            z ="Q4"
            list_d.append(z)
            grouped_df['quarter']= pd.DataFrame(list_d,columns=['ch'])
    i+=1
    return   grouped_df
```

```python
def cohort_months(grouped_df):
    #dict_month={"1":"jan","2":"feb","3":"march","4":"april","5":"may","6":"june","7":"july",
                #"8":"aug","9":"sep","10":"oct","11":"nov","12":"dec"}
    first_qt=[1,2,3]
    second_qt=[4,5,6]
    third_qt=[7,8,9]
    forth_qt=[10,11,12]
    list_d=[]
    for i in range(0,len(grouped_df)):
        if grouped_df.iloc[i]['first_order_month'] in first_qt:
            x ="Q1"
            list_d.append(x)
        elif grouped_df.iloc[i]['first_order_month'] in second_qt:
            y ="Q2"
            list_d.append(y)
        elif grouped_df.iloc[i]['first_order_month'] in third_qt:
            z ="Q3"
            list_d.append(z)
        else:
            z ="Q4"
            list_d.append(z)
            grouped_df['cohort_by_firstpurchase']= pd.DataFrame(list_d,columns=['ch'])
    i+=1
    return   grouped_df
```
